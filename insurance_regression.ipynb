```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Creting a scatter plot of age by cost of insurance

# Create our dataframe
ins_df = pd.read_csv('insurance.csv')
# View the data, make sure all is in order and familiarise myself with data
print(ins_df.head())

# Create an array of x values (independent variable) (the second colon makes it a vertical vector)
x1 = ins_df.loc[:,:'age'].values
# Create an array of y values (dependant variable)
y1 = ins_df.loc[:,'charges'].values

# Create our scatter plot with out defined x and y values
plt.scatter(x1, y1)
# Label the x axis
plt.xlabel('Age')
# Label the y axis
plt.ylabel('Insurance charges')
# Add a title
plt.title("Cost of Insurance by age")
# Print the plot
plt.show()

# The below commented out commands were just for me to check the correct data was being outputted.

# print(ins_df['age'].max())
# print(ins_df['age'].min())
# print(ins_df['charges'].max())
# print(ins_df['charges'].min())
```

```python
# Create the linear regression variable
ins_model = LinearRegression()
# Apply this model to our data
ins_model.fit(x1,y1)
```

```python
# x value to test (age) to determine insurance cost
test_value_x = [[80]]

# Print the prediction based on an age outside of the dataset
print(f"If the individual's age is {test_value_x}, we can approximate the insurance cost to be {ins_model.predict(test_value_x)}")

# Create our new predicted y axis based on the linear regression prediction
y1_pred = ins_model.predict(x1).reshape(-1,1)

    # The below is outputting the original scatter graph with the line of best fit

# Plot the original scatter graph from earlier
plt.scatter(x1,y1,)
# Plot the line of best fit with a distinct colour
plt.plot(x1,y1_pred,color = 'g')
# Label x axis
plt.xlabel('Age')
# Label y axis
plt.ylabel('Insurance charges')
# Add a title
plt.title("Cost of Insurance by age with line of best fit")
# Display the plot
plt.show()

    # The below is outputting an extended graph based on the predicted insurance charge

# Add our test value to the x axis
x2 = np.append(x1, test_value_x).reshape(-1,1)
# New y axis to accomodate the new value added
y2_pred = ins_model.predict(x2).reshape(-1,1)

# Plot the original scatter graph from earlier
plt.scatter(x1,y1,)
# Plot the line of best fit with a distinct colour
plt.plot(x2,y2_pred, color='black')
# Label x axis
plt.xlabel('Age')
# Label y axis
plt.ylabel('Insurance charges')
# Add a title
plt.title("Cost of Insurance by age with line of best fit - extended")
# Display the plot
plt.show()
```
